```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Import Data

```python
filename = "Water_Level_Field_Sheet.xlsx"

p1_16_02 = pd.read_excel(filename, skiprows = 7, sheet_name=0,
                        names = ["date", "time", "well_name", "depth", "MSL","comments"])
p2_16_02 = pd.read_excel(filename, skiprows = 7, sheet_name=1,
                        names = ["date", "time", "well_name", "depth", "MSL","comments"])
p3_16_02 = pd.read_excel(filename, skiprows = 7, sheet_name=2,
                        names = ["date", "time", "well_name", "depth", "MSL","comments"])


p1_18_01 = pd.read_excel(filename, skiprows = 7, sheet_name=3, nrows = 8,
                        names = ["date", "time", "well_name", "depth", "MSL","comments"])
p2_18_01 = pd.read_excel(filename, skiprows = 7, sheet_name=4, nrows = 8,
                        names = ["date", "time", "well_name", "depth", "MSL","comments"])
p3_18_01 = pd.read_excel(filename, skiprows = 7, sheet_name=5, nrows = 8,
                        names = ["date", "time", "well_name", "depth", "MSL","comments"])

p1_16_02["time"] = [0, 1, 2, 3, 4, 5]
p2_16_02["time"] = [0, 1, 2, 3, 4, 5]
p3_16_02["time"] = [0, 1, 2, 3, 4, 5]

p1_18_01["time"] = [0, 1, 2, 3, 4, 5, 6, 7]
p2_18_01["time"] = [0, 1, 2, 3, 4, 5, 6, 7]
p3_18_01["time"] = [0, 1, 2, 3, 4, 5, 6, 7]
```

```python
# define well length

p1_16_elevation = 1.209
p2_16_elevation = 1.214
p3_16_elevation = 1.217

p1_18_elevation = 1.58
p2_18_elevation = 1.579
p3_18_elevation = 1.581
```

```python
# Calculate head levels

p1_16_02["head"] = p1_16_elevation - p1_16_02["depth"]
p2_16_02["head"] = p2_16_elevation - p2_16_02["depth"]
p3_16_02["head"] = p3_16_elevation - p3_16_02["depth"]

p1_18_01["head"] = p1_18_elevation - p1_18_01["depth"]
p2_18_01["head"] = p2_18_elevation - p2_18_01["depth"]
p3_18_01["head"] = p3_18_elevation - p3_18_01["depth"]
```

```python
p2_18_01
```

```python
plt.figure()
fig, ax = plt.subplots(3, 2, figsize = (12, 10))

ax[0, 0].plot(p1_16_02["time"], p1_16_02["head"])
ax[0, 0].set_title("P1 16-02")
ax[0, 0].set_ylabel("Head Level (m)")
ax[0, 0].set_xlabel("Time (hrs)")

ax[1, 0].plot(p2_16_02["time"], p2_16_02["head"])
ax[1, 0].set_title("P2 16-02")
ax[1, 0].set_ylabel("Head Level (m)")
ax[1, 0].set_xlabel("Time (hrs)")

ax[2, 0].plot(p3_16_02["time"], p3_16_02["head"])
ax[2, 0].set_title("P3 16-02")
ax[2, 0].set_ylabel("Head Level (m)")
ax[2, 0].set_xlabel("Time (hrs)")

ax[0, 1].plot(p1_18_01["time"], p1_18_01["head"])
ax[0, 1].set_title("P1 18-01")
ax[0, 1].set_ylabel("Head Level (m)")
ax[0, 1].set_xlabel("Time (hrs)")

ax[1, 1].plot(p2_18_01["time"], p2_18_01["head"])
ax[1, 1].set_title("P2 18-01")
ax[1, 1].set_ylabel("Head Level (m)")
ax[1, 1].set_xlabel("Time (hrs)")

ax[2, 1].plot(p3_18_01["time"], p3_18_01["head"])
ax[2, 1].set_title("P3 18-01")
ax[2, 1].set_ylabel("Head Level (m)")
ax[2, 1].set_xlabel("Time (hrs)")

fig.tight_layout()
fig.show()
```
