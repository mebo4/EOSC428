```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.interpolate import griddata
```

### Import Data

```python
filename = "Water_Level_Field_Sheet.xlsx"
filename_p = "Vadose_Zone_Field_Sheets.xlsx"

p1_16_02 = pd.read_excel(filename, skiprows = 7, sheet_name=0,
                        names = ["date", "time", "well_name", "depth", "MSL","comments"])
p2_16_02 = pd.read_excel(filename, skiprows = 7, sheet_name=1,
                        names = ["date", "time", "well_name", "depth", "MSL","comments"])
p3_16_02 = pd.read_excel(filename, skiprows = 7, sheet_name=2,
                        names = ["date", "time", "well_name", "depth", "MSL","comments"])


p1_18_01 = pd.read_excel(filename, skiprows = 7, sheet_name=3, nrows = 8,
                        names = ["date", "time", "well_name", "depth", "MSL","comments"])
p2_18_01 = pd.read_excel(filename, skiprows = 7, sheet_name=4, nrows = 8,
                        names = ["date", "time", "well_name", "depth", "MSL","comments"])
p3_18_01 = pd.read_excel(filename, skiprows = 7, sheet_name=5, nrows = 8,
                        names = ["date", "time", "well_name", "depth", "MSL","comments"])

# tensiometer data
tens = pd.read_excel(filename_p, skiprows = 18, nrows = 5,
                     names = ["date", "time", "p_shallow", "h_eq_shallow", "Ls", "Ls_plus_Zps",
                             "p_deep", "h_eq_deep", "Ld", "Ld_plus_Zpd", "other"])

tens
```

### Calculate Hydraulic Head From Tensiometer
$$h = \frac{\psi}{\rho_w g} + z$$

```python
tens["time"] = [0, 3.5, 17.2]
tens
```

```python
elevation = 1.2 #m
z_deep = elevation -0.2 #m
z_shallow = elevation -0.16 #m
rho_water = 1000  # kg/m^3
g = 9.81 # m/s^2
```

```python
h_shallow = tens["p_shallow"] / (rho_water * g) + z_shallow
h_deep =  tens["p_deep"] / (rho_water * g) + z_deep
```

```python
fig, ax = plt.subplots(1, 2, figsize = (12, 4))

ax[0].plot(tens["time"], h_shallow)
ax[0].set_title("Hydraulic Head (Shallow Tensiometer)")
ax[0].set_xlabel("Time (hrs)")
ax[0].set_ylabel("Hydraulic Head")

ax[1].plot(tens["time"], h_deep)
ax[1].set_title("Hydraulic Head (Deep Tensiometer)")
ax[1].set_xlabel("Time (hrs)")
ax[1].set_ylabel("Hydraulic Head")

plt.tight_layout()
```

### Plot Well Locations

```python
filename_loc = "Geodetic_Data_2020.xlsx"
loc = pd.read_excel(filename_loc,
                   names = ["pt", "description", "E", "N", "z", "other"])

plt.figure(figsize = (4, 6))
plt.scatter(loc["E"], loc["N"])
plt.scatter(loc[4:7]["E"], loc[4:7]["N"], c = 'r', label = "(top)16-02")
plt.scatter(loc[15:18]["E"], loc[15:18]["N"], c = 'r', label = "(bottom)18-01")
plt.xlim(4982, 4996)
plt.ylim(4750, 4940)
plt.xlabel("Easting (m)")
plt.ylabel("Northing (m)")
plt.title("Well Locations")
plt.legend()

plt.tight_layout()
```

```python
E = np.array(loc["E"])
N = np.array(loc["N"])
Z = np.array(loc["z"])

grid_E, grid_N = np.meshgrid(
    np.linspace(E.min(), E.max(), 100),
    np.linspace(N.min(), N.max(), 100)
)

grid_Z = griddata((E, N), Z, (grid_E, grid_N), method='linear')

plt.figure(figsize=(6, 6))
contour = plt.contourf(grid_E, grid_N, grid_Z, levels=20, cmap='jet')
plt.colorbar(contour, label='Z (m)')
plt.xlabel('E (m)')
plt.ylabel('N (m)')
plt.xlim(4982, 4996)
plt.ylim(4750, 4940)
plt.title('Contour Plot of Elevation')
plt.scatter(loc["E"], loc["N"], c = 'k')
plt.scatter(loc[4:7]["E"], loc[4:7]["N"], c = 'r', label = "(top)16-02")
plt.scatter(loc[15:18]["E"], loc[15:18]["N"], c = 'r', label = "(bottom)18-01")
plt.legend()
# plt.scatter(E, N, c=Z, edgecolor='k', cmap='jet')  
```
